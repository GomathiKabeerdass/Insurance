```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
x['sex'] = le.fit_transform(x['sex'])
x['sex'].unique()
```

```python
x['region'] = le.fit_transform(x['region'])
x['region'].unique()
```

```python
#Standardscalar
from sklearn.preprocessing import StandardScaler
s=StandardScaler().fit_transform(x)
```

```python
s
```

```python
X = s
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.4,random_state=316)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
std_reg= LinearRegression().fit(x_train,y_train)
std_reg
```

```python
g = std_reg.predict(x_test)
g
```

```python
from sklearn.metrics import mean_absolute_error
k = std_reg.predict(x_test)
mean_absolute_error(k,y_test)
```

```python
from sklearn.metrics import mean_squared_error
a = std_reg.predict(x_test)
mean_squared_error(a,y_test)
```

**Finding MAE and MSE VALUE**

```python
from sklearn.metrics import r2_score
r2_score(g,y_test)
```
